#Fórmula da velocidade

```
def calculaVelMedia(deltaS, deltaT):
  velMedia = (deltaS/deltaT)
  return velMedia
```

## Teste Serviço Nuvem:Velocidade Media

```
print("Vm=" , calculaVelMedia(30, 15))
```

# Fórmula da aceleração

```
def calculaAceleracao(deltaV, deltaT):
  aceleração = (deltaV / deltaT)
  return aceleracao
```

```
print("a =", calculaAceleracao(20, 10))
```

# Fórmula do Movimento Uniforme

```
def movimentoUniforme(s0, v, t):
  s = s0 + (v * t)
  return s
```

```
print ("Posição final =", movimentoUniforme(10, 5, 2))
```

# Função Horária do Deslocamento

```
def funcaoHorariaDeslocamento(s0, v0, t, a):
  s = s0 + (v0 * t) + (a * t**2) / 2
  return s
```

#Teste de Função: Horaria de Deslocamento

```
print("S = ",funcaoHorariaDeslocamento(10, 5, 2, 2))
```
